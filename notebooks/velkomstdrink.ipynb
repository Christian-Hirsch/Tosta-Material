# Machine Learning Taxonomy

**Yann LeCun**

>Most of human and animal learning is unsupervised learning. If intelligence was a cake, __unsupervised learning__ would be the cake, __supervised learning__ would be the icing on the cake, and __reinforcement learning__ would be the cherry on the cake.


<img src="images/cake.jpg" alt="Drawing" style="width: 500px;"/>

## Unsupervised Learning

* AI learns __arbitrary structures in data__
* __Examples__ image generation, anonymization
* __Algorithms__ Clustering, Auto-Encoders, Generative Adversarial Networks


<img src="images/styleTrans.png" alt="Drawing" style="width: 1000px;"/>
https://research.googleblog.com/2016/10/supercharging-style-transfer.html

## Supervised Learning

* Machine learns via __manually annotated data__
* __Examples__ Image recognition, online ads, recommender systems
*  __Algorithms__ tree methods, deep learning, logistic regression, SVM


<img src="images/ilsvrc.png" alt="Drawing" style="width: 1000px;"/>
http://www.image-net.org/challenges/LSVRC/2014

## Reinforcement Learning

* Agent interacts with environment -- learning by sparse _reward signals_ ar
* __Examples__ AlphaGo, Atari, robotics
* __Algorithms__ Q-learning, policy-based methods, actor-critic approaches
